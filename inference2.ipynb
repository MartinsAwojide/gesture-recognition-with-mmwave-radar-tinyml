```python
label = ["right","stop"]

def inference(model, data):
    data = pd.read_csv(data, usecols=[3,5,6])
    data = (data - data.mean())/(data.std())
    data_ = data[:67].values.reshape(1,-1)
    output = np.argmax(model.predict(data_))
    pred = label[output]
    return pred 
    
```

```python
result = inference(model, "ei/arm_to_right/gesture_9.csv")
print(result)
```

```python
result = inference(model, "ei/arm_to_right/gesture_2.csv")
print(result)
```

```python
for i in range(1, 10): 
    result = inference(model, f"data/arm_to_right/gesture_{i}.csv")
    print(result)
```
